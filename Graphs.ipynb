```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ['SEP','Last','MA','MC','MV','CLAP']
data = np.array([[51.2,85.1,85.4,80.1,84.2,84.9],[33.8,-0.2,-0.5,4.8,0.7]],dtype=object)
bar_colors = ['gold', 'tab:blue', 'orchid', 'tab:orange', 'tab:green', 'tab:brown']
```

```python
fig, ax = plt.subplots()
ax.bar(x,data[0],color=bar_colors,label=x)
ax.set_title('In Distribution Performance')
ax.set_ylabel('AUC')
plt.show()
```
