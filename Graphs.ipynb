```python
from matplotlib.lines import Line2D

x = ['city','city','city','city','player','player','player','player','movie','movie','movie','movie']
ticks = range(len(x))
y1 = np.array([85.3,85.4,74.7,82.8
               ,52.0,55.9,52.3,51.7
               ,60.6,52.7,55.4,55.9])
bar_colors = ['gold', 'tab:blue',  'orchid', 'tab:orange', 'gold', 'tab:blue',  'orchid', 'tab:orange', 'gold', 'tab:blue',  'orchid', 'tab:orange']
# bar_labels = ['Last','Last','Last','MA','MA','MA','MC','MC','MC','MV','MV','MV']
legend_labels = [Line2D([0], [0], color='gold', lw=4, label='Last'),
                Line2D([0], [0], color='tab:blue', lw=4, label='MA'),
                Line2D([0], [0], color='orchid', lw=4, label='MC'),
                Line2D([0], [0], color='tab:orange', lw=4, label='MV')]

fig, axs = plt.subplots(1,3)

ax = axs[0]
bars = ax.bar(ticks,y1,color=bar_colors)
plt.sca(ax)
plt.xticks(ticks, x, rotation=90)
ax.set_title('Testing on City')
ax.set_ylabel('AUC')
ax.set_xlabel('Train data')
fig.legend(handles=legend_labels,title='Probing Strategy')
for i in [5,8]:
    bar = bars[i]
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                 f'*', ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np

# Define a larger font size
large_font_size = 14

# Data for the first subplot
x1 = ['city','city','city','city','', 'player','player','player','player','', 'movie','movie','movie','movie']
ticks1 = range(len(x1))
y1 = np.array([85.3,85.4,74.7,82.8, 0,
                 52.0,55.9,52.3,51.7, 0,
                 60.6,52.7,55.4,55.9])
bar_colors1 = ['gold', 'tab:blue',  'orchid', 'tab:orange', 'white',
               'gold', 'tab:blue',  'orchid', 'tab:orange', 'white',
               'gold', 'tab:blue',  'orchid', 'tab:orange']

# Data for the second subplot
x2 = ['player','player','player','player','', 'city','city','city','city','','movie','movie','movie','movie']
ticks2 = range(len(x2))
y2 = np.array([82.1,80.6,72.1,80.1, 0,
                 58.0,53.5,46.6,54.9, 0,
                 64.5,66.9,66.9,60.9])
bar_colors2 = ['gold', 'tab:blue',  'orchid', 'tab:orange', 'white',
               'gold', 'tab:blue',  'orchid', 'tab:orange', 'white',
               'gold', 'tab:blue',  'orchid', 'tab:orange']

# Data for the third subplot
x3 = ['movie','movie','movie','movie','', 'city','city','city','city','','player','player','player','player']
ticks3 = range(len(x3))
y3 = np.array([90.1,91.2,89.7,90.9, 0,
                 63.8,61.4,59.6,65.1, 0,
                 60.4,61.5,54.6,64.0])
bar_colors3 = ['gold', 'tab:blue',  'orchid', 'tab:orange', 'white',
               'gold', 'tab:blue',  'orchid', 'tab:orange', 'white',
               'gold', 'tab:blue',  'orchid', 'tab:orange']

legend_labels = [Line2D([0], [0], color='gold', lw=4, label='Last'),
                 Line2D([0], [0], color='tab:blue', lw=4, label='MA'),
                 Line2D([0], [0], color='orchid', lw=4, label='MC'),
                 Line2D([0], [0], color='tab:orange', lw=4, label='MV')]

fig, axs = plt.subplots(1, 3, figsize=(20, 7)) # Increased figure height for better text visibility

# First subplot
ax1 = axs[0]
bars1 = ax1.bar(ticks1, y1, color=bar_colors1)
ax1.set_xticks(ticks1)
ax1.set_xticklabels(x1, rotation=90, fontsize=large_font_size) # Increased x-axis label size
ax1.set_title('Testing on City', fontsize=large_font_size) # Increased title size
ax1.set_ylabel('AUC', fontsize=large_font_size) # Increased y-axis label size
ax1.set_xlabel('Train data', fontsize=large_font_size) # Increased x-axis title size
ax1.tick_params(axis='y', labelsize=large_font_size) # Increased y-axis tick size
for i in [6, 10]:
    bar = bars1[i]
    ax1.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'*', ha='center', va='bottom', fontsize=large_font_size + 2) # Increased asterisk size

# Second subplot
ax2 = axs[1]
bars2 = ax2.bar(ticks2, y2, color=bar_colors2)
ax2.set_xticks(ticks2)
ax2.set_xticklabels(x2, rotation=90, fontsize=large_font_size) # Increased x-axis label size
ax2.set_title('Testing on Player', fontsize=large_font_size) # Increased title size
ax2.set_ylabel('AUC', fontsize=large_font_size) # Increased y-axis label size
ax2.set_xlabel('Train data', fontsize=large_font_size) # Increased x-axis title size
ax2.tick_params(axis='y', labelsize=large_font_size) # Increased y-axis tick size
for i in [5, 11, 12]: # Place asterisks on the 6th, 12th, and 13th bars
    bar = bars2[i]
    ax2.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'*', ha='center', va='bottom', fontsize=large_font_size + 2) # Increased asterisk size

# Third subplot
ax3 = axs[2]
bars3 = ax3.bar(ticks3, y3, color=bar_colors3)
ax3.set_xticks(ticks3)
ax3.set_xticklabels(x3, rotation=90, fontsize=large_font_size) # Increased x-axis label size
ax3.set_title('Testing on Movie', fontsize=large_font_size) # Increased title size
ax3.set_ylabel('AUC', fontsize=large_font_size) # Increased y-axis label size
ax3.set_xlabel('Train data', fontsize=large_font_size) # Increased x-axis title size
ax3.tick_params(axis='y', labelsize=large_font_size) # Increased y-axis tick size
for i in [8, 13]: # Place asterisks on the 9th and 14th bars
    bar = bars3[i]
    ax3.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'*', ha='center', va='bottom', fontsize=large_font_size + 2) # Increased asterisk size

fig.legend(handles=legend_labels, title='Probing Strategy', loc='upper center', bbox_to_anchor=(0.5, 0), ncol=4, fontsize=large_font_size, title_fontsize=large_font_size)
plt.tight_layout(rect=[0, 0.05, 1, 0.95]) # Adjust layout to make space for the legend
plt.show()
```
