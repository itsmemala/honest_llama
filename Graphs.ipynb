```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy as np

data = np.array([[22.4  ,5.5], [62.9 ,32.6], [77.   ,2.9], [43.8  ,5.4], [54.7  ,4.5], [65.5 ,13.8]])
labels = ['L-7B TQA', 'L-7B STR','A-7B TQA', 'A-7B STR','V-7B TQA', 'V-7B STR']

# Number of groups (labels)
n_groups = len(labels)

# Data for each category
h_to_nh = data[:, 0]  # First column for %H->NH
nh_to_h = data[:, 1]  # Second column for %NH->H

# Set up bar width and positions
bar_width = 0.35
index = np.arange(n_groups) # The x locations for the groups

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 8)) # Adjust figure size for better readability

# Plotting the bars
bar1 = ax.bar(index - bar_width/2, h_to_nh, bar_width, label='Ratio of %H->NH', color='#1f77b4', alpha=0.9)
bar2 = ax.bar(index + bar_width/2, nh_to_h, bar_width, label='Ratio of %NH->H', color='#ff7f0e', alpha=0.9)

# Add labels and title
# ax.set_xlabel('Categories', fontsize=12)
ax.set_ylabel('Percentage (%)', fontsize=17)
ax.tick_params(axis='y', labelsize=17)
# ax.set_title('Percentage of H->NH and NH->H Transitions', fontsize=14)
ax.set_xticks(index)
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=17) # Rotate labels for better fit
ax.legend(fontsize=17)

# Add a grid for better readability
# ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of the bars
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=15)

autolabel(bar1)
autolabel(bar2)

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```
